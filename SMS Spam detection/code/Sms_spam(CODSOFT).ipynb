```python
import pandas as pd
df=pd.read_csv('spam.csv',encoding='latin-1')
df
```

```python
df.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
df.columns=['spam/ham','sms']
df
```

```python
df['spam/ham'].value_counts()
```

```python
df['spam/ham']=df['spam/ham'].map({'spam':1,'ham':0})
df
```

```python
sms=df.sms
sms
```

```python
sms=sms.str.replace("[^a-zA-Z0-9 ]","")
sms.head()
```

```python
from nltk.stem import SnowballStemmer
from nltk import TweetTokenizer
from nltk.corpus import stopwords
import nltk
tk=TweetTokenizer()
stemmer=SnowballStemmer('english')
nltk.download('stopwords')
stop=set(stopwords.words('english'))
```

```python
sms=sms.apply(lambda line:[stemmer.stem(token.lower()) for token in tk.tokenize(line)])
sms=sms.apply(lambda x:" ".join(x))
sms.head()
```

```python
sw=stopwords.words('english')
sw
```

```python
sms=sms.apply(lambda line:[token for token in tk.tokenize(line) if token not in sw]).apply(lambda token:" ".join(token))
sms.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
x=vectorizer.fit_transform(sms)
print(x)
```

```python
x.shape
```

```python
y=df['spam/ham']
y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
from sklearn.naive_bayes import MultinomialNB
nb=MultinomialNB()
from sklearn.svm import SVC
sv=SVC()
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
from sklearn.metrics import accuracy_score
```

```python
models=[nb,sv,lr]
for model in models:
  print("++++++++++++",model,"+++++++++++++")
  model.fit(x_train,y_train)
  y_pred=model.predict(x_test)
  print(accuracy_score(y_test,y_pred))
```

```python
x_new="Free entry in 2 a wkly comp to win FA Cup final tkts 21st May 2005. Text FA to 87121 to receive entry question(std txt rate)T&C's apply 08452810075over18's"
y_new=sv.predict(vectorizer.transform([x_new]))
if y_new==0:
  print("ham")
else:
  print("spam")
```
